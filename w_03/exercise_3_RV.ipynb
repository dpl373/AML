```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

a)

```python
x = [np.array([1,1]) ,np.array([-1, -1]) ,np.array([-1, 1]) ,np.array([1, -1])]
y = np.array([1, 1, -1, -1])
df = pd.DataFrame({"x": x, "y" : y})
```

```python
# Define colors based on labels
colors = {1: 'r', -1: 'b'}

# Create a figure and a 3D subplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Extract x, y, and z values
x = df['x'].apply(lambda x: x[0])
y = df['x'].apply(lambda x: x[1])
z = df['y']

# Plot the data points
ax.scatter(x, y, z, c=[colors[zi] for zi in z], marker='o', s=100)

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Label')

# Show the plot
plt.show()
```

```python
oxm = []
for x in df['x']:
   sqrt_x1 = np.sqrt(2) * x[0]
   sqrt_x2 = np.sqrt(2) * x[1]
   sqrt_x12 = np.sqrt(2) * x[0] * x[1]
   sq_x12 = x[0]**2 * x[1] ** 2

   mat = np.array([1, sqrt_x1, sqrt_x2, sqrt_x12, sq_x12]).T
   oxm.append(mat)
```
